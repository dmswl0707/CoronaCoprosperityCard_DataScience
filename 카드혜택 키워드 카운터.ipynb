```
import pandas as pd
import csv

#불러올 파일의 경로를 filename 변수에 저장
filename = '/content/drive/My Drive/test/allbenefits_combined.csv'

#pandas read_csv로 불러오기
data = pd.read_csv(filename)
del data['Unnamed: 0']
del data['Unnamed: 0.1']
data = data.fillna('')
print(data)
```

```
datacolumns = list(data.columns)
print(type(datacolumns))
```

```
for column in datacolumns:
    if column != 'name':
        keywordcount = data[column].str.split(expand=True).stack().value_counts()
        print('*'+column+'*')
        print(keywordcount)
        print('------------------')
```
