```python
import pandas as pd
import numpy as np
```

```python
seoul = pd.read_csv('seoul2019-2020.csv')
```

```python
seoul2019=seoul[seoul['기준_년_코드']==2019]
seoul2020=seoul[seoul['기준_년_코드']==2020]
```

```python
seoul.head(3)
```

# 2019,2020년 업종별 매출 합치기

```python
seoul2019_group = seoul2019.groupby('서비스_업종_코드_명')[['당월_매출_금액','점포수']].sum()
```

```python
seoul2019_group.sort_values('당월_매출_금액',ascending = False)
```

```python
seoul2020_group = seoul2020.groupby('서비스_업종_코드_명')[['당월_매출_금액','점포수']].sum()
```

```python
seoul2020_group.sort_values('당월_매출_금액',ascending = False)
```

# 업종별 매출감소율 확인

```python
service2019=seoul2019['서비스_업종_코드_명'].unique().tolist()
service2020=seoul2020['서비스_업종_코드_명'].unique().tolist()
```

```python
service=list(set(service2019) & set(service2020))
```

```python
new_data=[] # 업종 이름과 증감률값을 저장할 리스트 생성

for i in service: # 업종별로 계산
    value_2019=int(seoul2019_group[seoul2019_group.index==i]['당월_매출_금액'])
    value_2020=int(seoul2020_group[seoul2020_group.index==i]['당월_매출_금액'])
    
    rate = (value_2020-value_2019)/value_2019  # 증감률계산
    new_data.append([i,rate])

df = pd.DataFrame(new_data,columns=['서비스_업종_코드명','매출증감률'])   #데이터 프레임 생성
df
```

```python
df.sort_values('매출증감률',ascending = True)
```
