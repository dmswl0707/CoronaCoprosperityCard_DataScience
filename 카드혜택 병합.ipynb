```
import pandas as pd
import csv

#불러올 파일의 경로를 filename 변수에 저장
filename = '/content/drive/My Drive/test/Top100_card_key_benefits.csv'

#pandas read_csv로 불러오기
data = pd.read_csv(filename)
data = data.fillna('')
print(data)
```

```
#간편결제 합치기
data['간편결제혜택'] = data['APP']+data['PAYCO']+data['간편결제']+data['네이버페이']
del data['APP']
del data['PAYCO']
del data['간편결제']
del data['네이버페이']
```

```
#공과금 합치기
data['공과금혜택'] = data['공과금']+data['공과금/렌탈']
del data['공과금']
del data['공과금/렌탈']
```

```
#항공 합치기
data['항공혜택'] = data['PP']+data['공항']+data['공항라운지']+data['공항라운지/PP']+data['대한항공']+data['아시아나항공']+data['항공권']+data['라운지키']+data['항공마일리지']
del data['PP']
del data['공항']
del data['공항라운지']
del data['공항라운지/PP']
del data['대한항공']
del data['아시아나항공']
del data['항공권']
del data['라운지키']
del data['항공마일리지']
```

```
#차량 합치기
data['차량혜택'] = data['자동차']+data['자동차/하이패스']+data['정비']+data['주유']+data['주유소']
del data['자동차']
del data['자동차/하이패스']
del data['정비']
del data['주유']
del data['주유소']
```

```
#쇼핑 합치기
data['쇼핑혜택'] = data['대형마트']+data['도서']+data['드럭스토어']+data['모든가맹점']+data['백화점']+data['소셜커머스']+data['쇼핑']+data['아울렛']+data['온라인쇼핑']+data['해외']+data['해외이용']
del data['대형마트']
del data['도서']
del data['드럭스토어']
del data['모든가맹점']
del data['백화점']
del data['소셜커머스']
del data['쇼핑']
del data['아울렛']
del data['온라인쇼핑']
del data['해외']
del data['해외이용']
```

```
#생활혜택
data['생활혜택'] = data['교육/육아']+data['병원']+data['병원/약국']+data['뷰티/피트니스']+data['생활']+data['여행/숙박']+data['영화']+data['영화/문화']+data['테마파크']+data['경기관람']
del data['교육/육아']
del data['병원']
del data['병원/약국']
del data['뷰티/피트니스']
del data['생활']
del data['여행/숙박']
del data['영화']
del data['영화/문화']
del data['테마파크']
del data['경기관람']
```

```
#통신혜택
data['통신혜택'] = data['SKT']+data['통신']
del data['SKT']
del data['통신']
```

```
#음식혜택
data['음식혜택'] = data['SSM']+data['마트/편의점']+data['배달앱']+data['베이커리']+data['일반음식점']+data['점심']+data['카페']+data['카페/디저트']+data['패밀리레스토랑']+data['패스트푸드']+data['편의점']+data['푸드']
del data['SSM']
del data['마트/편의점']
del data['배달앱']
del data['베이커리']
del data['일반음식점']
del data['점심']
del data['카페']
del data['카페/디저트']
del data['패밀리레스토랑']
del data['패스트푸드']
del data['편의점']
del data['푸드']
```

```
#교통혜택
data['교통혜택'] = data['교통']+data['대중교통']+data['택시']+data['기차']
del data['교통']
del data['대중교통']
del data['택시']
del data['기차']
```

```
#카드혜택
data['카드혜택'] = data['멤버십포인트']+data['무이자할부']+data['선택형']+data['적립']+data['카드사']+data['캐시백']+data['하이브리드']+data['할인']
del data['멤버십포인트']
del data['무이자할부']
del data['선택형']
del data['적립']
del data['카드사']
del data['캐시백']
del data['하이브리드']
del data['할인']
```

```
#금융혜택
data['금융혜택'] = data['금융']+data['수수료우대']+data['은행사']
del data['금융']
del data['수수료우대']
del data['은행사']
```

```
#기타혜택
data['기타혜택'] = data['기타']+data['바우처']+data['프리미엄']+data['프리미엄 서비스']
del data['기타']
del data['바우처']
del data['프리미엄']
del data['프리미엄 서비스']
```
