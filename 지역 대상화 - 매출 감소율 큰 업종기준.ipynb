 매출감소율이 큰 업종이 많이 분포하는 지역에 혜택 부여 (점포수 기준은 2020년 2분기)

```python
import pandas as pd
import numpy as np

seoul = pd.read_csv('seoul2019_2020.csv')
seoul2020=seoul[(seoul['기준_년_코드']==2020) & (seoul['기준_분기_코드']==2)]
```

```python
seoul2020
```

```python
seoul2020 = seoul2020[['상권_코드_명','서비스_업종_코드_명','점포수']]
seoul2020
```

```python
seoul2020.iloc[0][2]
```

```python
group = seoul2020[seoul2020['서비스_업종_코드_명']=='컴퓨터및주변장치판매']
group['점포수'].max()
```

```python
group[group['점포수']==group['점포수'].max()]
```

```python
service_name_list = ['컴퓨터및주변장치판매','화장품','커피-음료','패스트푸드점']
new_list=[]


for i in range(0,len(service_name_list)):
    group = seoul2020[seoul2020['서비스_업종_코드_명']==service_name_list[i]]
    max_g = group[group['점포수']==group['점포수'].max()]
    new_list.append(max_g)

new_list
```

```python
pd.concat([new_list[0],new_list[1],new_list[2],new_list[3]])
```
