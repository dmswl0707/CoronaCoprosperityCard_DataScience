# 분기별 업종 매출 순위

```python
import pandas as pd
import numpy as np
import os
```

```python
seoul2019= pd.read_csv('seoul2019.csv',encoding='euc-kr')
seoul2020=pd.read_csv('seoul2020.csv',encoding='euc-kr')
```

# 분기별로 자르기

```python
seoul2019_1 = seoul2019[(seoul2019['기준_년_코드']==2019) & (seoul2019['기준_분기_코드']==1)]
seoul2019_2 = seoul2019[(seoul2019['기준_년_코드']==2019) & (seoul2019['기준_분기_코드']==2)]
```

```python
seoul2020_1 = seoul2020[(seoul2020['기준_년_코드']==2020) & (seoul2020['기준_분기_코드']==1)]
seoul2020_2 = seoul2020[(seoul2020['기준_년_코드']==2020) & (seoul2020['기준_분기_코드']==2)]
```

# 업종별로 매출 확인하기

# 19년도 1분기

```python
# 업종 전체매출 순위
seoul2019_1_sum = seoul2019_1.groupby('서비스_업종_코드_명')[['당월_매출_금액','당월_매출_건수','점포수']].sum()
seoul2019_1_sum = seoul2019_1_sum.sort_values(by=['당월_매출_금액'],ascending=False)
seoul2019_1_sum
```

```python
# 한 가게당 평균 매출 순위
seoul2019_1_sum['평균_매출_금액'] = seoul2019_1_sum['당월_매출_금액']/seoul2019_1_sum['점포수'] # 가게의 평균 매출을 보기위해 총 매출을 점포수로 나눠주었다
seoul2019_1_mean = seoul2019_1_sum.sort_values(by=['평균_매출_금액'],ascending=False)
seoul2019_1_mean
```

```python
#종합 - 각 매출상위 10개 업종을 리스트에 저장
seoul2019_1_list_sum = []
seoul2019_1_list_sum=seoul2019_1_sum.index.values[:10]
print('< 2019년도 1분기 전체 매출상위 업종 > \n',seoul2019_1_list_sum)

#평균 - 각 매출 상위 10개의 업종을 리스트에 저장
seoul2019_1_list_mean = []
seoul2019_1_list_mean=seoul2019_1_mean.index.values[:10]
print('\n< 2019년도 1분기 평균 매출상위 업종 > \n',seoul2019_1_list_mean)
```
